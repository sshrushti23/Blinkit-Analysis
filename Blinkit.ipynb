##Data Cleaning


```python
df['Item Fat Content']=df['Item Fat Content'].replace('LF','Low Fat')
```

```python
df['Item Fat Content']=df['Item Fat Content'].replace('low fat','Low Fat')
df['Item Fat Content']=df['Item Fat Content'].replace('reg','Regular')
```

```python
print(df['Item Fat Content'].unique())
```

##Buisness Requirements

```python
#Total Sales

total_sales=df['Total Sales'].sum()

##Average Sales

average_sales=df['Total Sales'].mean()

#No of item sold
no_of_items=df['Total Sales'].count()

##Average Rating

avg_ratings=df['Rating'].mean()

print(f"Total  Sales: {total_sales}")
print(f"Average Sales: {average_sales}")
print(f"No of Items: {no_of_items}")
print(f"Average Rating: {avg_ratings}")
```

##Charts Requirement


##Total Sales by Fat Content

```python
sales_by_fat=df.groupby('Item Fat Content')['Total Sales'].sum()
plt.pie(sales_by_fat,labels=sales_by_fat.index,autopct='%1.1f%%',startangle=90)
plt.title('Total Sales by Fat Content')
plt.show()
```

##Total Sales by Item Type

```python
sales_by_type=df.groupby('Item Type')['Total Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
bars = plt.bar(sales_by_type.index,sales_by_type.values)
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

for bar in bars:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.2f}',ha='center',va='bottom')

plt.show()
```

##Fat Content by Outlet For Total Sales

```python
grouped=df.groupby(['Outlet Location Type','Item Fat Content'])['Total Sales'].sum().unstack()
grouped=grouped[['Regular','Low Fat']]
ax= grouped.plot(kind='bar',figsize=(8,5),title="Outlet Tier by Item Fat Content")
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
```

##Total Sales by Outlet Establishment

```python
sales_by_year=df.groupby('Outlet Establishment Year')['Total Sales'].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index,sales_by_year.values,marker='o',linestyle='-')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')

for x,y in zip (sales_by_year.index,sales_by_year.values):
  plt.text(x,y,f'{y:.2f}',ha='center',va='bottom')

plt.show()
```

##Sales by Outlet Size

```python
sales_by_outlet_size=df.groupby('Outlet Size')['Total Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_outlet_size,labels=sales_by_outlet_size.index,autopct='%1.1f%%',startangle=90)
plt.title('Sales by Outlet Size')
plt.tight_layout()
plt.show()
```

##Total Sales by Location

```python
sales_by_location=df.groupby('Outlet Location Type')['Total Sales'].sum().reset_index()
sales_by_location=sales_by_location.sort_values('Total Sales',ascending=False)

plt.figure(figsize=(4,4))
ax=sns.barplot(x='Total Sales',y='Outlet Location Type',data=sales_by_location,palette='viridis')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')
plt.title('Total Sales by Outlet Location Type')
plt.tight_layout()
plt.show()
```
